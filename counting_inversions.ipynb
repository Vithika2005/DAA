```python
import random
import csv

def generate_csv(filename, num_students, num_subjects):
    """Generates a CSV file with student IDs and random course codes.

    Args:
        filename: The name of the CSV file to create.
        num_students: The number of students to generate.
        num_subjects: The number of subjects to generate course codes for.
    """

    with open(filename, 'w', newline='') as csvfile:
        fieldnames = ['StudentID'] + [f'Subject{i}' for i in range(1, num_subjects + 1)]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for i in range(1, num_students + 1):
            row = {'StudentID': i}
            for j in range(1, num_subjects + 1):
                row[f'Subject{j}'] = random.randint(100, 999)
            writer.writerow(row)

def count_inversions(row):
    """Counts the number of inversions in a row.

    An inversion occurs when two elements in the row are out of order.

    Args:
        row: A list of integers representing the elements in the row.

    Returns:
        The number of inversions in the row.
    """

    inversions = 0
    for i in range(len(row) - 1):
        for j in range(i + 1, len(row)):
            if row[i] > row[j]:
                inversions += 1
    return inversions

# Generate the CSV file
generate_csv('student_data.csv', 100, 6)

# Read the CSV file and count inversions
with open('student_data.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        student_id = row['StudentID']
        course_codes = [int(row[f'Subject{i}']) for i in range(1, 7)]
        inversions = count_inversions(course_codes)
        print(f"Student {student_id} has {inversions} inversions.")

        # Group students by inversion count
        if inversions == 0:
            print("Students with 0 inversions:", student_id)
        elif inversions == 1:
            print("Students with 1 inversion:", student_id)
        elif inversions == 2:
            print("Students with 2 inversions:", student_id)
        elif inversions == 3:
            print("Students with 3 inversions:", student_id)
```

```python
import csv
import os

# Function to count the number of inversions in a given array (student's course codes) using brute force
def count_inversions(arr):
    inv_count = 0
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] > arr[j]:
                inv_count += 1
    return inv_count

# Function to handle negative test cases and process each student's course codes
def process_students(csv_file):
    # Check if the file path is valid
    if not os.path.isfile(csv_file):
        raise FileNotFoundError(f"Error: Invalid file path '{csv_file}'.")

    # Dictionary to store the count of students having 0, 1, 2, 3 inversions
    inversion_counts = {0: 0, 1: 0, 2: 0, 3: 0}
    student_inversions = []  # To store each student's id and their inversion count
    student_ids = set()      # To track unique StudentIds
    row_number = 1           # To keep track of the row number in the CSV

    with open(csv_file, newline='') as file:
        reader = csv.DictReader(file)

        for row in reader:
            row_number += 1
            student_id = row['StudentID']

            # Check for missing StudentId and skip the row if it's missing
            if not student_id:
                print(f"Warning: Missing StudentId at row {row_number}. Skipping this row.")
                continue

            # Check for duplicate StudentIds
            if student_id in student_ids:
                print(f"Warning: Duplicate StudentId '{student_id}' found at row {row_number}.")
            else:
                student_ids.add(student_id)

            try:
                # Try to extract and convert course codes to integers
                course_codes = [
                    int(row['Subject1']),
                    int(row['Subject2']),
                    int(row['Subject3']),
                    int(row['Subject4']),
                    int(row['Subject5']),
                    int(row['Subject6'])
                ]
            except ValueError:
                # If conversion fails, print an error and continue
                print(f"Error: Invalid data at row {row_number}. All subject columns must contain integer values.")
                continue
            except KeyError:
                # If any subject is missing, print an error
                print(f"Error: Missing data at row {row_number}. Ensure all subject columns are present.")
                continue

            # Count inversions for this student's course codes using brute force method
            inversions = count_inversions(course_codes)

            # Append the student id and inversion count to the list
            student_inversions.append((student_id, inversions))

            # Update the inversion count summary (only for 0, 1, 2, 3 inversions)
            if inversions in inversion_counts:
                inversion_counts[inversions] += 1

    return student_inversions, inversion_counts

# Main function to handle file input and output
def main():
    csv_file = 'C:/Users/Vithika SURVE/OneDrive/Desktop/PYTHON PROJECTS/VITHIKA PYTHON/student_data.csv'  # Replace with the path to your CSV file
    
    try:
        student_inversions, inversion_counts = process_students(csv_file)
        
        # Output each student's ID and their inversion count
        print("\nStudent ID and their Count Inversions:")
        for student_id, inversions in student_inversions:
            print(f"Student ID: {student_id}, Inversions: {inversions}")
        
        # Output the summary of students with 0, 1, 2, 3 inversions
        print("\nSummary:")
        print(f"Students with 0 inversions: {inversion_counts[0]}")
        print(f"Students with 1 inversion: {inversion_counts[1]}")
        print(f"Students with 2 inversions: {inversion_counts[2]}")
        print(f"Students with 3 inversions: {inversion_counts[3]}")
    except FileNotFoundError as e:
        print(e)

# Run the main function
if __name__ == "__main__":
    main()
```
