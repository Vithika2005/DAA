```python
import csv
import time

def read_csv_data(filename):
    """Reads data from a CSV file and returns a list of tuples."""
    try:
        with open(filename, 'r') as csvfile:
            reader = csv.reader(csvfile)
            header = next(reader)  # Read header row
            data = []
            for row in reader:
                if len(row) != len(header):
                    raise ValueError("Missing data in row: {}".format(row))
                data.append((int(row[0]), row[1], row[2], int(row[3]), int(row[4]), int(row[5]), int(row[6])))
    except FileNotFoundError:
        raise ValueError("File path does not exist")

    for row in data:
        if any(value < 0 for value in row[3:7]):
            raise ValueError("Negative values found in salary columns")

    return data

def calculate_net_salary(basic_salary, housing_allowance, transportation_allowance, medical_allowance):
    # Replace this with your actual tax calculation logic
    # Consider factors like income tax slabs, professional tax, etc.
    total_deductions = basic_salary * 0.1  # Assuming 10% deductions for simplicity
    net_salary = basic_salary + housing_allowance + transportation_allowance + medical_allowance - total_deductions
    return net_salary

def find_min_max_divide_and_conquer(arr):
    """Finds the minimum and maximum elements using the divide and conquer approach."""
    if len(arr) == 1:
        return arr[0], arr[0]

    mid = len(arr) // 2
    left_min, left_max = find_min_max_divide_and_conquer(arr[:mid])
    right_min, right_max = find_min_max_divide_and_conquer(arr[mid:])

    return min(left_min, right_min), max(left_max, right_max)

def find_min_max_linear(arr):
    """Finds the minimum and maximum elements using the linear approach."""
    min_val = max_val = arr[0]
    for num in arr:
        if num < min_val:
            min_val = num
        if num > max_val:
            max_val = num
    return min_val, max_val

def main():
    filename = 'C:/Users/Vithika SURVE/OneDrive/Desktop/PYTHON PROJECTS/VITHIKA PYTHON/employee_data(14).csv'#Replace with your CSV file path

    try:
        data = read_csv_data(filename)

        # Filter out rows with zero salaries
        filtered_data = [row for row in data if all(value > 0 for value in row[3:7])]

        # Calculate net salaries for filtered data
        net_salaries = [calculate_net_salary(row[3], row[4], row[5], row[6]) for row in filtered_data]

        # Find minimum and maximum net salaries
        min_net_salary = min(net_salaries) if net_salaries else None
        max_net_salary = max(net_salaries) if net_salaries else None

        # Write net salaries to a new CSV file
        output_filename = 'employee_data_with_net_salary.csv'
        with open(output_filename, 'w', newline='') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['Employee ID', 'First Name', 'Last Name', 'Basic Salary', 'Housing Allowance', 'Transportation Allowance', 'Medical Allowance', 'Total Salary', 'Net Salary'])
            for row, net_salary in zip(data, net_salaries):
                writer.writerow(row + (net_salary,))

        print("Net Salaries (excluding zero salaries):")
        if min_net_salary is not None and max_net_salary is not None:
            print("  Minimum Net Salary:", min_net_salary)
            print("  Maximum Net Salary:", max_net_salary)
        else:
            print("  No valid salaries found after excluding zero values.")
    except ValueError as e:
        print("Error:", e)

if __name__ == "__main__":
    main()
```
